**Q1.** Download *user.csv*  from D2L. The entries in *user*  represent movie reviewers.

Add a column to the dataframe called *agegroup*, where the value of *agegroup* is a label based on the following table. Using this table, you can create relevant intervals (i.e. left-open and right-closed intervals).

Infant: 0-1 <br>
Toddler: 1-3 <br>
Children: 3-12 <br>
Teen: 12-19  <br>
Young-adult: 19-25 <br> 
Adult: 25 - 65 <br>
Senior: 65 - 110 <br>

So, someone aged 13 will be labeled 'Teen'


```python
import pandas as pd
import numpy as np

u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
user = pd.read_csv('user.csv', sep='|', names=u_cols)

#defining an array which contains the values of separation, binning 

bins = [0,1, 3,12,19,25,65,110]

#defining an array which contains the labels for intervals 

grpNames = ['Infant', 'Toddler', 'Children', 'Teen', 'Young-adult', 'Adult', 'Senior']


#replacing intervals with meaningful labels and gets a series  
categories = pd.cut(user['age'], bins, labels=grpNames) 


#adding ageGroup to the DataFrame 
user = pd.DataFrame(user, columns = ['user_id', 'age', 'sex', 'occupation', 'zip_code', 'agegroup'])

#adding a series to DataFrame 
user['agegroup'] = categories

user



```

**Q2.2**  Print the movie ids of the top rated movies (rating > 4) that are rated by either Teens or Seniors.

 

```python
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
rating = pd.read_csv('rating.csv', sep='\t', names=r_cols)

# write your code below

# Q2.1
newFrame = pd.merge(user, rating)

# Q2.2
#producing dataframe with rows that have Teen or Senior values for the agegroup values 
newF1 = newFrame[newFrame['agegroup'].isin(['Teen', 'Senior'])]; 

#producing dataframe with rows that have a rating > 4
newF2 = newF1.loc[newF1['rating'] > 4]
newF2['movie_id']




```
